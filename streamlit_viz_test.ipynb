```python
#Imports
import pandas as pd
```

```python
df = pd.read_csv("./data/sabot_events.log")
```

```python
df = df[df["type"] == "Swapped"] 
df
```

```python
df_data = pd.DataFrame(df[20:], columns=["time", "type", "utilityTokenTotalSupply", "balanceOfBaseToken", "balanceOfTargetToken"])
df_data
# df_data_clean = df_data.dropna()
# df_data_clean
```

# CLOG Token Price Timeline

```python
df_data["CLOG_Price"] = ( df_data["balanceOfBaseToken"] + df_data["balanceOfTargetToken"] ) / df_data["utilityTokenTotalSupply"]
df_data
```

```python
df_data_clean = df_data[["time", "CLOG_Price"]].dropna()
df_data_clean
```

```python
df2 = df_data.copy()
df2
```

```python
df2["balanceOfTargetToken"] = 50000000000000000000
df2
```

```python
df_data = pd.concat([df_data, df2])
df_data.reset_index(inplace = True)
df_data
```

```python
df_data["CLOG_Price"] = ( df_data["balanceOfBaseToken"] + df_data["balanceOfTargetToken"] ) / df_data["utilityTokenTotalSupply"]
df_data
```

```python
df_data.plot(
x= "level_0",
y="CLOG_Price")
```

```python
df_data.plot(
x = "level_0",
y = "utilityTokenTotalSupply"
)
```
